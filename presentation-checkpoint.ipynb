### A simply supported beam of 5m is subjected to clockwise moment of 10kN-m at a distance of 2m from the left end.

```python
import numpy as np
import matplotlib.pyplot as plt
def beam_analysis():
    LENGTH = 5  #m
    BENDING_MOMENT = 10    # kNm
    
    #calculate reactions at A and B
    R_b= BENDING_MOMENT/ LENGTH
    R_a= -(R_b)
    
    
    #calculate shear forces
    F_a= -2 #kN
    F_b= -2 #kN
    F_c= -2 #kN
    
    #calculate bending moments
    M_a=0
    M_b=0
    M_c1= R_a*2 #On the left side of C
    M_c2=M_c1 + BENDING_MOMENT # On the right side of C
    
    positions=[0,2,5]
    shears=[F_a,F_b,F_c]
    moments=[M_a,M_b,M_c1,M_c2]
    
    print("Reactions:")
    print(f"at A:{R_a}kN")
    print(f"at B:{R_b}kN")
    
    print("Shear Forces:")
    for i in range(len(positions)):
        print(f"{positions[i]}m:{shears[i]}kN")
    
    print("Bending Moment:")
    print(f"at A:{M_a}kNm")
    print(f"at B:{M_b}kNm")
    print(f"On the left side of C :{M_c1}kNm")
    print(f"On the right side of C :{M_c2}kNm")
    

beam_analysis()
    
```

### Shear Force Diagram

```python
fig,ax = plt.subplots()
x = [0,5]
y = [-2,-2]
ax.fill_between(x,y,color="blue",alpha=0.5,hatch="||||")
plt.xlabel('Position  (m)')
plt.ylabel('Shear Force  (kN)')
plt.title('Shear Force Diagram')

plt.grid(True)
plt.show()
```

### Bending Moment Diagram

```python
fig,ax = plt.subplots()
x = [0,0,2,2,2,5,5]
y = [0,0,-4,0,6,0,0]
ax.fill_between(x,y,color="blue",alpha=0.5,hatch="||||")
plt.xlabel('Position  (m)')
plt.ylabel('Bending Moment  (kNm)')
plt.title('Bending Moment Diagram')

plt.grid(True)
plt.show()
```
